```python
import pandas as pd
df=pd.read_csv('dirtydata.csv')
x=df['Calories'].mean()
print(x)
df['Calories'].fillna(x,inplace=True)
print(df)
```

```python
import pandas as pd
df=pd.read_csv('dirtydata.csv')
x=df['Gender'].mode()
print(x[0])
df['Gender'].fillna(x[0],inplace=True)
print(df)
```
